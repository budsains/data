## Machine learning
Machine Learning (ML) merupakan pendekatan atau metode untuk menjadikan mesin memiliki kemampuan untuk belajar (learning) tanpa harus diprogram (diberikan perintah) secara eksplisit.

ML merupakan bagian dari artifisial inteligence (AI) yang berkembang dari suatu konsep pemrograman mesin yang sebelumnya diberikan rules atau aturan tertentu menjadi rules yang tidak harus diberikan secara eksplisit, namun kita biarkan mesin tersebut untuk mempelajari pola-pola dari data yang ada.

input output program.png

Sebagai ilustrasi, sebuah robot diprogram atau diberikan rules untuk berhenti ketika ada lubang, menyala jika ada getaran. Robot bekerja sesuai dengan rules yang telah kita berikan. Dalam perkembangannya, kita lelah untuk selalu memberikan rules-rules tersebut, maka kita ingin si mesin dapat belajar sendiri untuk membaca pola-pola kondisi yang dihadapi. Dalam hal ini kita tidak memprogram mesin tersebut, namun kita memanfaatkan data sehingga robot bisa mengenali (input) dan mempelajari polanya sehingga pada akhirnya bisa membuat output sendiri sesuai pola yang dipelajari (misal ada lubang robot berhenti).

# Preprocessor


```
import pandas as pd
from sklearn.model_selection import train_test_split

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
```

```
df = pd.read_csv("https://raw.githubusercontent.com/WiraDKP/supervised_learning/refs/heads/master/04%20-%20KNN%20%26%20Scikit-learn/data/titanic.csv", index_col="PassengerId")
df.columns
```

```
X = df.drop(columns=["Survived","Name", "Ticket", "Age", "Cabin"])
y = df.Survived

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify=y, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Pipeline




```
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder

```

```
numerical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', MinMaxScaler())
])
```

```
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder())
])
```

```
X_train.head()
```

```
preprocessor = ColumnTransformer([
    ('numerical', numerical_pipeline, ['SibSp','Parch','Fare']),
    ('categorical', categorical_pipeline, ['Pclass','Sex','Embarked'])
])
```

### pipeline

```
from sklearn.neighbors import KNeighborsClassifier
```

```
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', KNeighborsClassifier())
])
```

```
pipeline.fit(X_train, y_train)
```

```
pipeline.score(X_test, y_test)
```
